```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ---   ( ) ---
INITIAL_INVESTMENT = 6_000_000  #   ( +  + )
MORTGAGE_CITY_COST = 20_000_000 #     ()
YEARS = 20                      #  

#   ( %)
REAL_ESTATE_GROWTH = 0.05       #   
NATURE_CAPITAL_GROWTH = 0.10    #    (, , )
PASSIVE_INCOME_START = 3        #     (, )
PASSIVE_INCOME_YEARLY = 500_000 #     
INCOME_GROWTH = 0.07            #  

# ---   ---
years = np.arange(YEARS + 1)
city_capital = []
estate_capital = []
passive_income_accumulated = []

current_city_val = MORTGAGE_CITY_COST #  (,    ,   %  )
# :   .     .
#       3%,          .
city_net_asset = 4_000_000 #  
city_growth_rate = 0.03

current_estate_val = INITIAL_INVESTMENT
current_nature_val = 100_000 #    ()
total_income = 0

for y in years:
    # :  ,  
    city_net_asset = city_net_asset * (1 + city_growth_rate) + 500_000 #   
    city_capital.append(city_net_asset)
    
    # :
    # 1.  
    current_estate_val *= (1 + REAL_ESTATE_GROWTH)
    # 2.    (, )
    current_nature_val *= (1 + NATURE_CAPITAL_GROWTH)
    
    # 3. 
    if y >= PASSIVE_INCOME_START:
        yearly_inc = PASSIVE_INCOME_YEARLY * ((1 + INCOME_GROWTH) ** (y - PASSIVE_INCOME_START))
        total_income += yearly_inc
    
    #   
    total_estate_capital = current_estate_val + current_nature_val + total_income
    estate_capital.append(total_estate_capital)
    passive_income_accumulated.append(total_income)

# ---  ---
plt.figure(figsize=(12, 6))
plt.style.use('dark_background') # Hi-Tech 

plt.plot(years, np.array(estate_capital) / 1_000_000, label='  (Neuro-Estate)', color='#00ff99', linewidth=3, marker='o')
plt.plot(years, np.array(city_capital) / 1_000_000, label='   ()', color='#ff6666', linewidth=2, linestyle='--')

plt.fill_between(years, np.array(estate_capital) / 1_000_000, alpha=0.2, color='#00ff99')

plt.title(' :  vs   (20 )', fontsize=16, fontweight='bold', color='white')
plt.xlabel('', fontsize=12)
plt.ylabel(' ', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=12)

# 
plt.annotate(f' \n -', xy=(PASSIVE_INCOME_START, estate_capital[PASSIVE_INCOME_START]/1000000), 
             xytext=(PASSIVE_INCOME_START+2, estate_capital[PASSIVE_INCOME_START]/1000000 - 5),
             arrowprops=dict(facecolor='white', shrink=0.05), color='white')

plt.annotate(f': {estate_capital[-1]/1_000_000:.1f} ', xy=(YEARS, estate_capital[-1]/1000000), 
             xytext=(YEARS-5, estate_capital[-1]/1000000 + 5),
             arrowprops=dict(facecolor='#00ff99', shrink=0.05), fontsize=12, fontweight='bold', color='#00ff99')

plt.tight_layout()
plt.show()

print(f"---   ( {YEARS} ) ---")
print(f"    ( ): {city_capital[-1]:,.0f} .")
print(f"   Neuro-Estate ( + ): {estate_capital[-1]:,.0f} .")
print(f"   (): +{(estate_capital[-1] - city_capital[-1]):,.0f} .")
```
